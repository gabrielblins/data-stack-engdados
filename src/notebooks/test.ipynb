```python
pairs_url_results = []

for pair in pairs:
    url = f"https://api.cryptowat.ch/pairs/{pair}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        pairs_url_results.append(data)
    else:
        pairs_url_results.append(None)
```

```python
pairs_url_results
```

```python
symbol_exchanges_dict = {}

for item in pairs_url_results:
    symbol = item['result']['symbol']
    markets = item['result']['markets']
    exchanges = [market['exchange'] for market in markets]
    symbol_exchanges_dict[symbol] = exchanges
```
